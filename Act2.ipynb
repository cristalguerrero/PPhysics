```python
import matplotlib.pyplot as plt

class Particle:
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r

def generate_particles_equidistant(num_particles, space_size, particle_radius):
    particles = []
    spacing = (space_size - 2 * particle_radius) / (num_particles**0.5 - 1)  # Espaciado entre partculas
    num_per_row = int(num_particles**0.5)  # Nmero de partculas por fila

    for i in range(num_per_row):
        for j in range(num_per_row):
            if len(particles) < num_particles:
                x = particle_radius + i * spacing
                y = particle_radius + j * spacing
                particles.append(Particle(x, y, particle_radius))

    return particles

def plot_particles(particles, space_size):
    fig, ax = plt.subplots()
    ax.set_xlim(0, space_size)
    ax.set_ylim(0, space_size)
    ax.set_aspect('equal')

    for particle in particles:
        circle = plt.Circle((particle.x, particle.y), particle.r, color='blue', alpha=0.6)
        ax.add_artist(circle)

    plt.title(f"Equidistant Placement of {len(particles)} Particles")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Parmetros
    num_particles = 49       # Nmero de partculas (debe ser un cuadrado perfecto)
    space_size = 100         # Tamao del espacio (cuadrado de 100x100)
    particle_radius = 2      # Radio de cada partcula

    # Generar y graficar las partculas
    particles = generate_particles_equidistant(num_particles, space_size, particle_radius)
    plot_particles(particles, space_size)


import random
import matplotlib.pyplot as plt

class Particle:
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r

def generate_particles_random(num_particles, space_size, particle_radius):
    particles = []

    for _ in range(num_particles):
        # Genera coordenadas aleatorias dentro del espacio definido
        x = random.uniform(particle_radius, space_size - particle_radius)
        y = random.uniform(particle_radius, space_size - particle_radius)
        particles.append(Particle(x, y, particle_radius))

    return particles

def plot_particles(particles, space_size):
    fig, ax = plt.subplots()
    ax.set_xlim(0, space_size)
    ax.set_ylim(0, space_size)
    ax.set_aspect('equal')

    for particle in particles:
        circle = plt.Circle((particle.x, particle.y), particle.r, color='blue', alpha=0.6)
        ax.add_artist(circle)

    plt.title(f"Random Placement of {len(particles)} Particles")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Parmetros
    num_particles = 50      # Nmero de partculas
    space_size = 100         # Tamao del espacio (cuadrado de 100x100)
    particle_radius = 2      # Radio de cada partcula

    # Generar y graficar las partculas
    particles = generate_particles_random(num_particles, space_size, particle_radius)
    plot_particles(particles, space_size)
```
